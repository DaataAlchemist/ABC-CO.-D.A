# 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
T_count= df['Team'].value_counts()
per_split = T_count/ len(df) * 100
df1= pd.DataFrame({'Count': T_count, 'Percentage': per_split})
df1
```

# 2.Segregate the employees w.r.t different positions.

```python
import matplotlib.pyplot as plt

df1['Percentage'].plot.pie(figsize=(10,10), autopct='%1.1f%%', startangle=90, radius=1.2)
plt.title("Employees as per Team", fontsize=16, y=1.08)
plt.legend(bbox_to_anchor=(1.5, 1), loc='upper left')
plt.show()
```

```python
df4=df.groupby("Position")["Name"].nunique()
```

# 3.Find from which age group most of the employees belong to.

```python
age=df['Age'].value_counts().head(30)
age
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Age",data=df)
plt.title("Age Group Category")
plt.show()
```

```python
bins= [15,20,25,30,35,40]
labels=["15-20","21-25","26-30","31-35","36-40"]
df["Age_Group"]=pd.cut(df["Age"],bins=bins,labels=labels)
sns.countplot(x="Age_Group",data=df)
plt.title("Age Group Category")
plt.show()
```

# 4. Find out under which team and position, spending in terms of salary is high.

```python
df_ts=pd.DataFrame(df.groupby(["Team","Position"])["Salary"].sum())
df_ts
```

```python
df_ts.sort_values(by="Salary", ascending=False)
```
